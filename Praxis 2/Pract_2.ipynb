## Задание 1

Создадим вспомогательные функции для разделения списка республик случайным образом на 3 части и для сложения 2-мерного массива в одномерный (np.flatten() не подходит)

```python
republics = ['РСФСР', 'Украинская ССР', 'Белорусская ССР', 'Узбекская ССР',
          'Казахская ССР', 'Грузинская ССР', 'Азербайджанская ССР',
          'Литовская ССР', 'Молдавская ССР', 'Латвийская ССР', 'Киргизская ССР',
          'Таджикская ССР', 'Армянская ССР', 'Туркменская ССР', 'Эстонская ССР']

departs = ['Сельское хозяйство', 'Легкая промышленность', 'Тяжелая промышленность группы А',
           'Тяжелая промышленность группы Б', 'Военно промышленный комплекс',
           'Наука', 'Химическая промышленность']

def get_random_departs(departs):
    deps = departs.copy()
    r_deps = []
    for i in range(3):
        ra.shuffle(deps)
        
        size = ra.randint(1, len(deps) - 2 + i)
        sampl = deps[:size]
        
        if i == 2:
            sampl = deps
        
        r_deps.append(sampl)
        
        deps = deps[size:]
        
    return r_deps


def l_flatten(lst):
    temp_list = []
    for el in lst:
        try:
            temp_list.extend(el)
        except TypeError:
            temp_list.append(el)
    return temp_list
```

Создадим датафрейм с характеристиками для каждой республики и заполним его случайным образом отраслями, а также подсчитаем индексы развития республик в целом

```python
repubs = sorted(ra.sample(republics, ra.randint(3, 15)))
length = len(repubs)
data = [(repub, '', '', '', 0, 0) for repub in repubs]
data = pd.DataFrame(data, columns=['Название республики', 'Избыточно развитые', 'Сбалансированно развитые', 'Слабо развитые',
                             'Индекс избыточно развитых', 'Индекс слабо развитых'])
data
```

```python
keys = ['Избыточно развитые', 'Сбалансированно развитые', 'Слабо развитые']
for i in range(len(data)):
    deps = get_random_departs(departs)
    for j in range(3):
        data.at[i, keys[j]] = deps[j]
        if j == 0:
            data.at[i, 'Индекс избыточно развитых'] = len(deps[j])
        elif j == 2:
            data.at[i, 'Индекс слабо развитых'] = -len(deps[j])
data
```

```python
deps = [(dep, 0, 0, 0, 0) for dep in departs]
deps = pd.DataFrame(deps, columns=['Отрасль', 'Избыточно развитые', 'Сбалансированно развитые',
                                   'Слабо развитые', 'Общий индекс развития'])
deps
```

Подсчитаем количество отраслей в каждой категории и заполним таблицу с данными отраслей в соответствие с массивами из первой таблицы

```python
for key in keys:
    dep = l_flatten(data[key])
    for i in range(len(deps)):
        deps.at[i, key] = dep.count(deps.at[i, 'Отрасль'])
        
for i in range(len(deps)):
    dep = deps.iloc[i]
    deps.at[i, 'Общий индекс развития'] = dep['Избыточно развитые'] - dep['Слабо развитые']
deps
```

Создадим специальный фрейм для сбора статистики о пиковых отраслях.<br>Индекс развития сбалансированной отрасли не связан с индексами для развитой и отстающей отраслей, потому что индекс сбалансированности не помножен на свой коэффициент (0) для сохранения значения.<br>В некоторых полях отраслей может быть несколько из-за схожести показателей.

```python
stats = pd.DataFrame(columns=['Название отрасли', 'Индекс развития'])
desces = ['Самая развитая отрасль', 'Самая сбалансированная отрасль', 'Самая отстающая отрасль']
for key, desc in zip(keys, desces):
    dep = deps[deps[key] == deps[key].max()].to_dict('list')
    if key == 'Слабо развитые':
        indexes = -dep[key][0]
    else:
        indexes = dep[key][0]
    stats.loc[desc] = ', '.join(dep['Отрасль']), indexes
stats
```
