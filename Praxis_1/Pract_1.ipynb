## Задание 1

```python
def m_eval(roundoff=4):
    try:
        num_1, num_2 = map(d, input('Введите 2 числа через пробел: ').split())
        op = input('Введите один из операторов (+, -, /, *, **, %, **- для извлечения корня): ')
        if op in ['+', '-', '/', '*', '**', '%']:
            pass
        elif op == '**-':
            op = f'** (1/{num_2})'
            num_2 = ''
        else:
            raise ValueError
        return round(eval(f'{num_1} {op} {num_2}'), roundoff)
    except:
        return 'Error'
```

Тестирование на выполнение всех функций

```python
for _ in range(7):
    print(f'Result: {m_eval()}')
```

Тестирование на перехват ошибок

```python
for _ in range(10):
    print(f'Result: {m_eval()}')
```
