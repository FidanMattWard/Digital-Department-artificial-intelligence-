## Задание 1

Создадим вспомогательные функции для разделения списка республик случайным образом на 3 части и для сложения 2-мерного массива в одномерный (np.flatten() не подходит)<br>get_random_departs возвращает список из 3 списков, каждый из которых относится к соответствующей категории отраслей, ключи представлены ниже.

```python
republics = ['РСФСР', 'Украинская ССР', 'Белорусская ССР', 'Узбекская ССР',
          'Казахская ССР', 'Грузинская ССР', 'Азербайджанская ССР',
          'Литовская ССР', 'Молдавская ССР', 'Латвийская ССР', 'Киргизская ССР',
          'Таджикская ССР', 'Армянская ССР', 'Туркменская ССР', 'Эстонская ССР']

departs = ['Сельское хозяйство', 'Легкая промышленность', 'Тяжелая промышленность группы А',
           'Тяжелая промышленность группы Б', 'Военно промышленный комплекс',
           'Наука', 'Химическая промышленность']

def get_random_departs(departs):
    deps = departs.copy()
    r_deps = []
    for i in range(3):
        ra.shuffle(deps)
        
        size = ra.randint(1, len(deps) - 2 + i)
        sampl = deps[:size]
        
        if i == 2:
            sampl = deps
        
        r_deps.append(sampl)
        
        deps = deps[size:]
        
    return r_deps


def l_flatten(lst):
    temp_list = []
    for el in lst:
        try:
            temp_list.extend(el)
        except TypeError:
            temp_list.append(el)
    return temp_list
```

Создадим датафрейм с характеристиками для каждой республики и заполним его случайным образом отраслями, а также подсчитаем индексы развития республик в целом

```python
repubs = sorted(ra.sample(republics, ra.randint(3, 15)))
data = [(repub, '', '', '', 0, 0) for repub in repubs]
data = pd.DataFrame(data, columns=['Название республики', 'Избыточно развитые', 'Сбалансированно развитые', 'Слабо развитые',
                             'Индекс избыточно развитых', 'Индекс слабо развитых'])
data
```

```python
keys = ['Избыточно развитые', 'Сбалансированно развитые', 'Слабо развитые']
for i in range(len(data)):
    deps = get_random_departs(departs)
    for j in range(3):
        data.at[i, keys[j]] = deps[j]
data['Индекс избыточно развитых'] = data['Избыточно развитые'].apply(len)
data['Индекс слабо развитых'] = -data['Слабо развитые'].apply(len)
data
```

```python
deps = [(dep, 0, 0, 0, 0) for dep in departs]
deps = pd.DataFrame(deps, columns=['Отрасль', 'Избыточно развитые', 'Сбалансированно развитые',
                                   'Слабо развитые', 'Общий индекс развития'])
deps
```

Подсчитаем количество отраслей в каждой категории и заполним таблицу с данными отраслей в соответствие с массивами из первой таблицы

```python
for key in keys:
    dep = l_flatten(data[key])
    deps[key] = deps['Отрасль'].apply(lambda x: dep.count(x))
        
deps['Общий индекс развития'] = deps['Избыточно развитые'] - deps['Слабо развитые']
deps
```

Создадим таблицу для сбора статистики о пиковых отраслях.<br>Индекс развития сбалансированной отрасли не связан с индексами для развитой и отстающей отраслей, потому что индекс сбалансированности не помножен на свой коэффициент (0) для сохранения значения.<br>В некоторых полях отраслей может быть несколько из-за схожести показателей.

```python
stats = pd.DataFrame(columns=['Название отрасли', 'Индекс развития'])
desces = ['Самая развитая отрасль', 'Самая сбалансированная отрасль', 'Самая отстающая отрасль']
for key, desc in zip(keys, desces):
    dep = deps[deps[key] == deps[key].max()].to_dict('list')
    indx = dep[key][0] * (-1) ** (key == 'Слабо развитые')
    stats.loc[desc] = ', '.join(dep['Отрасль']), indx
stats
```

Статистика отраслей по показателям развития со всех стран

```python
deps[['Отрасль', 'Общий индекс развития']].sort_values('Общий индекс развития', ascending=False)
```

```python
def calc_coord(time, veloc, accel=1, x_0=0):
    return x_0 + veloc * time + accel * time ** 2 / 2


time, veloc = map(float, input('time, velocity: ').split())
print(calc_coord(time, veloc))

time, veloc, accel = map(float, input('time, velocity, acceleration: ').split())
print(calc_coord(time, veloc, accel))

time, veloc, accel, x_0 = map(float, input('time, velocity, acceleration, initial coordinate: ').split())
print(calc_coord(time, veloc, accel, x_0))

time, veloc, x_0 = map(float, input('time, velocity, initial coordinate: ').split())
print(calc_coord(time, veloc, x_0=x_0))
```

```python
products = ['котел', 'турбина', 'проволока', 'лента', 'шплинт',
            'шпонка', 'гвозди', 'шурупы', 'болты', 'винты',
            'дюбели', 'анкеры', 'гайки', 'шпильки', 'штифты',
            'шайбы', 'конфирматы', 'подшипники', 'талрепы', 'коуши']
```

```python
def get_unique(products):
    ra.shuffle(products)
    set_1 = set(ra.sample(products, ra.randint(3, len(products))))
    
    ra.shuffle(products)
    set_2 = set(ra.sample(products, ra.randint(3, len(products))))
    
    return set_1.symmetric_difference(set_2), set_1, set_2


diff, set_1, set_2 = get_unique(products)
print(*sorted(set_1))
print(*sorted(set_2))
print('Symmetric difference:')
print(*sorted(diff))
```
